```python
df = sex_age_df[(sex_age_df['Group'] == 'By Total') & (sex_age_df['State'] != 'United States') & (sex_age_df['Sex'] != 'All Sexes') & (sex_age_df['Age Group'] != 'All Ages')]
df.drop(['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month', 'Footnote'], axis = 1, inplace= True)
df  = df[~df['Age Group'].isin(['0-17 years', '18-29 years', '30-39 years', '40-49 years', '50-64 years'])]
```

```python
def fill_row(x):
    if(pd.isnull(x)):
        return np.random.randint(low = 1, high = 9)
    return x
def impute(series_):
    series_ = series_.apply(lambda x: fill_row(x))
    return series_
df = df.apply(lambda x: impute(x))
```

```python
def func(x):
    if x == "Under 1 year":
        return 0
    elif x == "85 years and over":
        return 85
    return x.split("-")[0]
df['Age Group'] = df['Age Group'].apply(lambda x : func(x))
```

```python
df.to_csv('../data/cdc_sex_age_state_filtered.csv')
```
