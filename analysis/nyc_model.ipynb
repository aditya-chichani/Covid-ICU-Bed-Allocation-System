```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
# prepare the lists for the model
X = nyc_daily_data[['CASE_COUNT_7DAY_AVG','ALL_CASE_COUNT_7DAY_AVG','HOSP_COUNT_7DAY_AVG','HOSPITALIZED_COUNT','CASE_COUNT']]
y = nyc_daily_data['ALL_DEATH_COUNT_7DAY_AVG'].tolist()[0:]
# date format is not suitable for modeling, let's transform the date into incrementals number starting from April 1st
starting_date = 106 # April 1st is the 37th day of the series
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
# # let's train our model only with data after the peak
# Instantiate Linear Regression
regr = RandomForestRegressor(max_depth=12, random_state=10)
regr.fit(X_train, y_train)
print("R2 score :", regr.score(X_test,y_test))
```
